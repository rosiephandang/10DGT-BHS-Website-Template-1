```python
# @title Code Cell #3
from bs4 import BeautifulSoup
from playwright.async_api import async_playwright

from PIL import Image
import numpy as np
from collections import Counter
import IPython.display as Idisplay
import io
from PIL import ImageDraw, ImageFont
import time
import string
from unidecode import unidecode

# DEFINITIONS -------------------

cookies = [{"name" : "_otwarchive_session",
            "value" : _otwarchive_session,
            "domain" : ".archiveofourown.org",
            "path" : "/"},

            {"name" : "cf_clearance",
            "value" : cf_clearance,
            "domain" : ".archiveofourown.org",
            "path" : "/"},

            {"name" : "__cf_bm",
            "value" : __cf_bm,
            "domain" : ".archiveofourown.org",
            "path" : "/"},

            {"name" : "_cfuvid",
            "value" : _cfuvid,
            "domain" : ".archiveofourown.org",
            "path" : "/"}]

if cf_clearance == "":
    cookies = [{"name" : "_otwarchive_session",
            "value" : _otwarchive_session,
            "domain" : ".archiveofourown.org",
            "path" : "/"},

            {"name" : "__cf_bm",
            "value" : __cf_bm,
            "domain" : ".archiveofourown.org",
            "path" : "/"},

            {"name" : "_cfuvid",
            "value" : _cfuvid,
            "domain" : ".archiveofourown.org",
            "path" : "/"}]


class Fic:
    def __init__(self, link, title, rating, authors, fandoms, language,
                 relationships, characters, freeforms, words):
        self.link = link
        self.title = title
        self.rating = rating
        self.authors = authors ## default is ['Anonymous']
        self.fandoms = fandoms
        self.relationships = relationships ## default of tags is empty array
        self.characters = characters
        self.freeforms = freeforms
        self.language = language
        self.words = words


    def __eq__(self, other):
        if isinstance(other, Fic):
            return self.link == other.link  # Assuming links are unique
        return False

    def __hash__(self):
        return hash(self.link)  # Ensuring that identical fics are treated as the same


async def get_page_html(page_url, cookies):
  async with async_playwright() as p:
    browser = await p.chromium.launch(headless=True)
    context = await browser.new_context(
        user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
        (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36",
        locale="en-US",
        viewport={'width': 1280, 'height': 720}
    )

    await context.add_cookies(cookies)
    page = await context.new_page()

    # await page.goto("https://archiveofourown.org/users/login")
    # await page.fill("#user_login", username)
    # await page.fill("#user_password", password)
    # await page.click("input[name='commit']:visible")

    tries = 0

    while True:
      try:
        await page.goto(page_url, timeout=15000)
        html = await page.content()

        if "SSL handshake failed" in html or "Cloudflare" in html or "Retry later" in html:
          raise Exception("Error page returned by AO3.")

        await browser.close()
        return html

      except Exception as e:
        print("Blocked by AO3, retrying after two minutes. Error message: ", e)
        time.sleep(120)
        tries+=1
        if (tries >= 3):
          await browser.close()
          raise Exception("Could not load page after three tries.")



def create_fic(work_soup):
  heading = work_soup.find("h4", class_="heading")
  link = "archiveofourown.org" + heading.find('a').get('href')
  title = heading.find('a').text

  rating = work_soup.find("a", class_="help symbol question modal modal-attached", title="Symbols key").text

  authors = []
  try:
      authors_list = work_soup.find("a", rel="author")
      for i in authors_list:
          authors.append(i.text)
  except:
      authors.append('Anonymous')

  fandoms = []
  fandoms_list = work_soup.find("h5", class_="fandoms heading").find_all("a", class_="tag")
  for i in fandoms_list:
      fandoms.append(i.text)

  relationships = []
  try:
      relationships_list = work_soup.find_all("li", class_="relationships" )
      for i in relationships_list:
          relationships.append(i.text.replace('\n', ''))
  except:
      pass

  characters = []
  try:
      characters_list = work_soup.find_all("li", class_="characters" )
      for i in characters_list:
          characters.append(i.text.replace('\n', ''))
  except:
      pass

  freeforms = []
  try:
      freeforms_list = work_soup.find_all("li", class_="freeforms" )
      for i in freeforms_list:
          freeforms.append(i.text.replace('\n', ''))
  except:
      pass

  try:
      words = int(work_soup.find("dd", class_= "words").text.replace(',', ''))
  except:
      print(f"Words for fic: {link} not counted due to processing error.")
      words = 0

  try: language = work_soup.find('dd', class_= "language").text
  except: language = ""

  return Fic(link, title, rating, authors, fandoms, language, relationships, characters, freeforms, words)




# DATA COLLECTION ------------------------

isDone = False

link = "https://archiveofourown.org/users/" + username + "/readings"


if num_pages == 0:
  try:
    page_html = await get_page_html(link+"?page=1", cookies)

    soup = BeautifulSoup(page_html, 'lxml')
    pages = soup.find("ol", attrs={"class": "pagination actions pagy"})
    num_pages = int(pages.find_all('li')[-2].text)

  except:
    num_pages = 1

# iterate through pages of history

fics = []
curr_page = start_page

while (not isDone) and curr_page <= num_pages:

  if curr_page > 1 and curr_page%10 == 1:
    print(f"Sleeping for {wait_time} seconds.")
    time.sleep(wait_time)

  page_url = link + f"?page={curr_page}"
  print("Opening page", curr_page)

  curr_page_html = await get_page_html(page_url, cookies)
  page_soup = BeautifulSoup(curr_page_html, 'lxml')
  work_soups = page_soup.find('ol', class_="reading work index group").find_all("li", role="article")
  visit_info = page_soup.find_all('h4', attrs={'class': "viewed heading"})

  for i in range(len(work_soups)): # iterate fics on page

    if "Marked for Later" in visit_info[i].text:
      continue # ignore if in marked for later

    if "Deleted work" in visit_info[i].text:
      continue # ignore if deleted

    if "Mystery Work" in work_soups[i].text:
      continue # ignore if mystery work

    to_exclude = False
    work_title = work_soups[i].find("h4", class_="heading").find('a').text
    for t in fics_to_exclude:
      if t.casefold() == work_title.casefold():
        to_exclude = True
        break

    work_authors = []
    try:
        authors_list = work_soups[i].find("a", rel="author")
        for j in authors_list:
            work_authors.append(j.text)
    except:
        work_authors.append('Anonymous')

    for auth in authors_to_exclude:
      if auth in work_authors:
        to_exclude = True
        break

    if to_exclude:
      continue

    #fic = create_fic(work_soups[i])
    #fics.append(fic)
    ## get year of visit
    year_last_viewed = int(visit_info[i].text[22:26])

    if year_last_viewed < wrapped_year:
      isDone = True
      break

    elif year_last_viewed > wrapped_year:
      continue

    else:
      fic = create_fic(work_soups[i])
      fics.append(fic)

  curr_page = curr_page+1

# DATA PROCESSING -------------------------------

total_fics = len(fics)
word_counts = [fic.words for fic in fics]
total_words = np.sum(word_counts)
longest_fic = fics[np.argmax(word_counts)]

def find_top(flattened_list, num):
  c = Counter(flattened_list)
  if len(flattened_list) > num:
    return c.most_common(num)
  else:
    return c.most_common()

def find_top_authors(fics, num):
  flat_authors = [author for fic in fics for author in fic.authors]
  filtered = [author for author in flat_authors if author not in ['orphan_account', 'Anonymous']]
  c = Counter(filtered)
  if len(filtered) > num:
    return c.most_common(num)
  else:
    return c.most_common()


top5Authors = find_top_authors(fics, 5)
favCharacters = find_top([character for fic in fics for character in fic.characters], 2)
top3Fandoms = find_top([fandom for fic in fics for fandom in fic.fandoms], 3)
top5Tags = find_top([tag for fic in fics for tag in fic.freeforms], 5)
favRating = find_top([fic.rating for fic in fics], 3)
favShips = find_top([relationship for fic in fics for relationship in fic.relationships], 10)

sorted_fics_wc = sorted(fics, key=lambda fic: fic.words, reverse=True)
top_5_by_wc = sorted_fics_wc[:5]
longest5Fics = [(fic.title, fic.words) for fic in top_5_by_wc]

print("Favorite authors:", top5Authors)
print("Favorite characters:", favCharacters)
print("Favorite fandoms:", top3Fandoms)
print("Favorite tags:", top5Tags)
print("Favorite rating:", favRating)
print("Favorite ships:", favShips)
print("Longest fics:", longest5Fics)

### Different comments
def commentRating(favR):
  if favR == 'Explicit':
    com = "No judgement."
  elif favR == 'Mature':
    com = "You're a little spicy."
  elif favR == 'Teen And Up Audiences':
    com = "Jesus is proud."
  elif favR == 'General Audiences':
    com = "We love to see it!"
  else: com = " "
  return com

def commentWords(wordsTot):
  if wordsTot < 50000:
    com= "Congrats, it seems that you touch grass."
  elif wordsTot < 120000:
    com= "That's a whole novel!"
  elif wordsTot < 250000:
    com= "That's almost the longest \nHarry Potter book!"
  elif wordsTot < 320000:
    com= "Look at that, that could be \na A Song of Ice and Fire book!"
  elif wordsTot < 450000:
    com= "You've almost finished \nthe Earthsea series!"
  elif wordsTot < 600000:
    com= "That's almost the equivalent \nof the Lord of the Ring series!"
  elif wordsTot < 1000000:
    com= "That's almost the equivalent \nof the Harry Potter series!"
  elif wordsTot < 1800000:
    com= "That's almost the equivalent \nof the Song of Ice & Fire series!"
  elif wordsTot > 1800000:
    com= f"Wow, that's about {wordsTot//365} words a day."

  return com

def commentLength(l):
  com = ""
  if l>= 1000 and l <= 5000:
    com = "You're more of a oneshot \nkind of person."
  elif l< 50000:
    com = "Hey, that's the \naverage length of a novella!"
  elif l< 100000:
    com = "That's around the length of \nof the first Harry Potter book!"
  elif l < 120000:
    com = "Good job, you've basically \nread a large novel!"
  elif l>120000:
    com = "This one is longer \nthan your average novel!"
  elif l>1000000:
    com = "Wow, that's one dedicated author."
  return com

def commentTrope(trope):
  if trope == 'Not Canon Compliant' or trope == 'Alternate Universe - Canon Divergence':
     com = "Canon? What canon?"
  elif trope == 'Fluff':
    com = "I see you're keeping your dentist busy."
  elif (trope == 'Angst' or trope == "Hurt No Comfort" or trope == "Whump" or trope == 'Death'
      or trope == 'Violence' or trope == 'Blood' or trope == 'Abuse' or trope == 'Hurt' or trope == 'Character Death'):
    com = "...Are you okay, friend?"
  elif (trope == 'Sexual Content' or trope == 'Sex' or trope =='Smut'
      or trope == 'Oral Sex' or trope == 'Anal Sex' or trope == 'Anal'
      or trope == 'BDSM' or trope == "Porn" or trope == "Bottoming"):
    com = "Bonk??"
  elif trope == 'Hurt/Comfort':
    com = "... emphasis on Comfort."
  elif trope == 'Humor' or trope == 'Crack' or trope == 'Light-Hearted':
     com = "Better laugh than cry, right?"
  elif trope == 'Established Relationship':
     com = "Ah yes, we love the domesticity."
  elif trope == 'Happy Ending':
      com = "We all like those."
  elif trope == 'One Shot':
     com = "Make it short!"
  elif trope == 'Slow Burn' or trope =='Pining' or trope == "Friends to Lovers":
     com = "I can smell the pine trees from here."
  elif trope == 'Happy Ending' or trope =='Angst with a Happy Ending':
     com = "We all need the happy ending."
  elif trope == 'Plot What Plot/Porn Without Plot':
     com = "Who needs plot?"
  elif trope == 'Fluff and Angst' or trope == "Fluff and Smut":
     com = "Because why not both?"
  elif trope == 'Enemies to Lovers':
    com = "Ohh, you wanna kiss me so bad."
  elif trope == 'Alternate Universe - College/University' or trope =="University" or trope =="Canon Compliant":
    com = "An entire world of fiction, and this is where you go."
  elif trope == 'Alpha/Beta/Omega Dynamics':
    com = "Hey, it's knot anything to be ashamed of."
  elif trope == 'Dead Dove: Do Not Eat':
    com = 'Close that fridge!'
  elif trope == 'Incest':
    com = "Sweet Home Alabama"
  else:
    com = "Never gets old."
  return com


## your comments:
com_wordsTot = commentWords(total_words)
com_rating = commentRating(favRating[0][0])
com_longest = commentLength(longest_fic.words)
com_trope = commentTrope(top5Tags[0][0])


def splitString(a, maxl):

  ret = [a]

  while len(ret[-1]) > maxl:

    curr_str = ret[-1]
    first_space = maxl
    while first_space > 0 and curr_str[first_space] != ' ':
      first_space = first_space-1

    if first_space == 0:
      ret[-1] = curr_str[:maxl-2] + '..\n'
      ret.append(curr_str[maxl-2:])
    else:
      ret[-1] = curr_str[:first_space] + '\n'
      ret.append(curr_str[first_space+1:])

  return "".join(ret)


def toEnglish(a):
  return unidecode(a)

def splitFandom(a):
  return a.replace(' | ', ' |\n', 1)

def splitShip(ship):
  if "/" in ship:
    i = ship.index('/')
    return toEnglish(ship[:i+1]) + '\n' + toEnglish(ship[i+1:])
  if "&" in ship:
    i = ship.index('&')
    return toEnglish(ship[:i+1]) + '\n' + toEnglish(ship[i+1:])

  return splitString(ship, 20)

def splitAU(AU):
  i = AU.index('-')
  return AU[:i+1] + '\n' + AU[i+1:]


fontBig = ImageFont.truetype("/content/AO3_wrapped_blank_images/LeagueSpartan-Bold.ttf", 150)
fontMed = ImageFont.truetype("/content/AO3_wrapped_blank_images/LeagueSpartan-Bold.ttf", 70)
fontSmall = ImageFont.truetype("/content/AO3_wrapped_blank_images/LeagueSpartan-Bold.ttf", 50)

fontMed_maxLetters = 19

### FIRST PIC -------------------------------------
image1 = Image.open('/content/AO3_wrapped_blank_images/1.png')
imgW, imgH = image1.size
d1 = ImageDraw.Draw(image1)
red = (151, 0, 0)

## ---------
text1 = str(total_fics)
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2], text1, font=fontBig)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 550), text1, font=fontBig, fill =red)

## ---------
text2 = str(total_words)
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2], text2, font=fontBig)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 900), text2, font=fontBig, fill =red)
text3 = "words"
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text3, font=fontMed)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 1030), text3, font=fontMed, fill =(0,0,0))
text3 = com_wordsTot
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text3, font=fontSmall)
w, h = right - left, top - bottom
d1.text((imgW/2, 1150), text3, font=fontSmall, fill =(0,0,0), anchor ='mm', align='center')

## ---------
text4 = longest_fic.title
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text4, font=fontMed)
w, h = right - left, top - bottom
if w>650:
  #split
  text4 = splitString(text4, fontMed_maxLetters)
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text4, font=fontMed)
w, h = right - left, top - bottom
d1.text((imgW/2, 1430 - h/2), text4, font=fontMed, fill = red, align='center', anchor = 'mm')

text5 = "by "
for author in longest_fic.authors[:-1]:
  text5 += author + ', '
text5 += longest_fic.authors[-1]

left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text5, font=fontMed)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 1600), text5, font=fontMed, fill = (0,0,0))

text6 = com_longest
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text6, font=fontSmall)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 1680), text6, font=fontSmall, fill = (0,0,0), align='center')

text7 = str(wrapped_year)
left, top, right, bottom = d1.textbbox( [imgW/2,imgH/2],text7, font=fontMed)
w, h = right - left, top - bottom
d1.text((imgW/2 - w/2, 280), text7, font=fontMed, fill=(0, 0, 0))



### SAVE IMAGE 1
image1.save("ao3_wrapped_pg_1.png", format='PNG')

### SECOND PIC -------------------------------------
image2 = Image.open('/content/AO3_wrapped_blank_images/2.png')
imgW, imgH = image2.size
d2 = ImageDraw.Draw(image2)

## ---------
text1 = toEnglish(top3Fandoms[0][0])
left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text1, font=fontMed)
w, h = right - left, top - bottom
while w > 850 and ' | ' in text1:
  text1 = splitFandom(text1)
  left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2],text1, font=fontMed)
  w, h = right - left, top - bottom
d2.text((imgW/2, 700-h/2), text1, font=fontMed, fill =(250, 250, 250), align = 'center', anchor = 'mm')

if len(top3Fandoms) > 1:
  text2 = toEnglish(top3Fandoms[1][0])
  left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text2, font=fontSmall)
  w, h = right - left, top - bottom
  while w > 850 and ' | ' in text2:
      text2 = splitFandom(text2)
      left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text2, font=fontSmall)
      w, h = right - left, top - bottom
  d2.text((imgW/2 - w/2, 930), text2, font=fontSmall, fill = red, align = 'center')
  h2 = h

if len(top3Fandoms) > 2:

  text3 = '&'
  left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
  w, h = right - left, top - bottom
  d2.text((imgW/2- w/2 , 1020 + h2/2+ 30), text3, font=fontSmall, fill = (0,0,0))

  text3 = toEnglish(top3Fandoms[2][0])
  left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
  w, h = right - left, top - bottom
  while w > 850 and ' | ' in text3:
    text3 = splitFandom(text3)
    left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
    w, h = right - left, top - bottom
  d2.text((imgW/2 - w/2 , 1050 + h2/2 + 50), text3, font=fontSmall, fill = red, align = 'center')

## ---------
text4 = toEnglish(favCharacters[0][0])
left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text4, font=fontMed)
w, h = right - left, top - bottom
if w > 620:
    text4 = splitString(text4, 19)
left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2],text4, font=fontMed)
w, h = right - left, top - bottom
d2.text((imgW/2 , 1350-h/2), text4, font=fontMed, fill =red, align = 'center', anchor = 'mm')

text5 = "who appears in " + str(favCharacters[0][1])+ " works"
left, top, right, bottom = d2.textbbox( [imgW/2,imgH/2], text5, font=fontSmall)
w, h = right - left, top - bottom
d2.text((imgW/2 - w/2, 1500), text5, font=fontSmall, fill = (0,0,0), align = 'center')


### SAVE RESULT FOR IMG 2
image2.save("ao3_wrapped_pg_2.png", format='PNG')


### THIRD PIC -------------------------------------
image3 = Image.open('/content/AO3_wrapped_blank_images/3.png')
imgW, imgH = image3.size
d3 = ImageDraw.Draw(image3)

## ---------
text1 = splitShip(favShips[0][0])
left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text1, font=fontMed)
w, h = right - left, top - bottom
d3.text((40, 320-h/2), text1, font=fontMed, fill =(0,0,0))

if len(favShips) > 1:

  text2 =splitShip(favShips[1][0])
  left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text2, font=fontSmall)
  w, h = right - left, top - bottom
  d3.text((380, 570-h/2), text2, font=fontSmall, fill =(0,0,0))

if len(favShips) > 2:

  fontSmaller = ImageFont.truetype("/content/AO3_wrapped_blank_images/LeagueSpartan-Bold.ttf", 35)
  text3 = splitShip(favShips[2][0])
  left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text3, font=fontSmaller)
  w, h = right - left, top - bottom
  d3.text((690, 780-h/2), text3, font=fontSmaller, fill =(0,0,0))

## ---------
text1 = str(favShips[0][1]) + " fics"
left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text1, font=fontSmall)
w, h = right - left, top - bottom
d3.text((100, 550-h/2), text1, font=fontSmall, fill =(250,250,250))

if len(favShips) > 1:

  text2 = str(favShips[1][1]) + " fics"
  left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text2, font=fontSmall)
  w, h = right - left, top - bottom
  d3.text((450, 760-h/2), text2, font=fontSmall, fill =(250,250,250))

if len(favShips) > 2:

  text3 = str(favShips[2][1]) + " fics"
  left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
  w, h = right - left, top - bottom
  d3.text((800, 930-h/2), text3, font=fontSmall, fill =(250,250,250))

## ---------
for i in range(len(top5Authors)):
  text = top5Authors[i][0]
  left, top, right, bottom = d3.textbbox( [imgW/2,imgH/2], text, font=fontMed)
  w, h = right - left, top - bottom
  d3.text((imgW/2 - w/2, 1400 + i*70), text, font=fontMed, fill =(0,0,0))

### SAVE RESULT FOR IMG 3
image3.save("ao3_wrapped_pg_3.png", format='PNG')


### FOURTH PIC -------------------------------------
image4 = Image.open('/content/AO3_wrapped_blank_images/4.png')
imgW, imgH = image4.size
d4 = ImageDraw.Draw(image4)

## ---------
text1 = top5Tags[0][0]
if '-' in text1:
  text1 = splitAU(text1)
elif '/' in text1:
  text1 = splitShip(text1)
elif len(text1) > 20:
   text1 = splitString(text1, 20)

left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text1, font=fontMed)
w, h = right - left, top - bottom
d4.text((650, 400-h/2), text1, font=fontMed, fill =red, anchor ="mm")

text = "with " + str(top5Tags[0][1]) + " works"
left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text, font=fontSmall)
w, h = right - left, top - bottom
d4.text((500, 500-h/2), text, font=fontSmall, fill =(0,0,0))

text2 = splitString(com_trope, 20)
left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text2, font=fontSmaller)
w, h = right - left, top - bottom
d4.text((670- w/2, 580-h/2), text2, font=fontSmaller, fill =(0,0,0))

## ---------

if len(top5Tags) > 1:

  text3 = top5Tags[1][0]
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
  w, h = right - left, top - bottom
  if '-' in text3:
    text3 = splitAU(text3)
  elif '/' in text3:
    text3 = splitShip(text3)
  elif len(text3) > 20:
    text3 = splitString(text3, 20)

  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text3, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((770, 880-h/2), text3, font=fontSmall, fill =red, anchor ="mm")

if len(top5Tags) > 2:

  text4 = top5Tags[2][0]
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text4, font=fontSmall)
  w, h = right - left, top - bottom
  if '-' in text4:
    text4 = splitAU(text4)
  elif '/' in text4:
    text4 = splitShip(text4)
  elif len(text4) > 20:
    text4 = splitString(text4, 20)

  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text4, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((700, 1170-h/2), text4, font=fontSmall, fill =red, anchor ="mm")

if len(top5Tags) > 3:

  text5 = top5Tags[3][0]
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text5, font=fontSmall)
  w, h = right - left, top - bottom
  if '-' in text5:
    text5 = splitAU(text5)
  elif '/' in text5:
    text5 = splitShip(text5)
  elif len(text5) > 20:
    text5 = splitString(text5, 20)

  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text5, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((600, 1420-h/2), text5, font=fontSmall, fill =red, anchor ="mm")

## ---------

if len(top5Tags) > 1:
  text = "with " + str(top5Tags[1][1]) + " works"
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((580, 960-h/2), text, font=fontSmall, fill =(0,0,0))

if len(top5Tags) > 2:

  text = "with " + str(top5Tags[2][1]) + " works"
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((500, 1250-h/2), text, font=fontSmall, fill =(0,0,0))

if len(top5Tags) > 3:

  text = "with " + str(top5Tags[3][1]) + " works"
  left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text, font=fontSmall)
  w, h = right - left, top - bottom
  d4.text((420, 1490-h/2), text, font=fontSmall, fill =(0,0,0))

## ---------
text6 = favRating[0][0]
if  text6 == 'Teen And Up Audiences':
  text6 = text6[:12] + "\n" +  text6[12:]

left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text6, font=fontMed)
w, h = right - left, top - bottom
d4.text((300- w/2, 1700-h/2), text6, font=fontMed, fill =(250,250,250))

text7 = "that's " + str(favRating[0][1]) + " works."
left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text7, font=fontSmall)
w, h = right - left, top - bottom
d4.text((570, 1750-h/2), text7, font=fontSmall, fill =(0,0,0))

text8 = com_rating
left, top, right, bottom = d4.textbbox( [imgW/2,imgH/2], text8, font=fontSmall)
w, h = right - left, top - bottom
d4.text((570, 1800-h/2), text8, font=fontSmall, fill =(0,0,0))


### SAVE RESULT FOR IMG 4------------------------------------

image4.save("ao3_wrapped_pg_4.png", format='PNG')

## SHOW IMG 1
buf = io.BytesIO()
image1.save(buf, format='PNG')
byte_im = buf.getvalue()
Idisplay.Image(byte_im)
```
